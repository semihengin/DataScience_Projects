```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plot
```

```python
df_platforms = pd.read_csv("MoviesOnStreamingPlatforms.csv")
df_netflix = pd.read_csv("netflix1.csv")
```

```python
sns.countplot(x='type', data=df_netflix)
plot.title("movie/series rates on netflix")
plot.show()

df_netflix.type.value_counts().plot(kind = 'pie', autopct = "%1.f%%")
plot.legend(labels=['Movie','TV Shows'], loc='lower left') 
plot.title("movie/series rates on netflix")
plot.show()
```

```python
# Load the Netflix dataset
df = df_netflix

# Counting the number of movies and TV series published on a yearly basis in the data set
years = df.groupby("release_year").size()

# Bar graf
plot.figure(figsize=(10,10))
plot.bar(years.index, years.values)
plot.xlabel("Years")
plot.ylabel("Number of Movies/Series")
plot.title("Number of Movies and Series Streamed on Netflix by Year")
plot.show()
```

```python
# Load the Netflix data set and skip missing values in the "country" column
df = pd.read_csv("netflix1.csv").dropna(subset=["country"])

# Count the number of movies and series released in each country
countries = df.groupby("country").size()

# Sort the countries by the number of movies and series in decreasing order
countries = countries.sort_values(ascending=False)

# Identify the countries that have less than 100 movies or series
other_countries = countries[countries <= 100]

# Identify the countries that have more than 100 movies or series
countries = countries[countries > 100]

# Create a bar chart
plot.bar(countries.index, countries.values, color="blue")

# Use the wrap function to display the country names properly
plot.xticks(rotation=90, ha="right", wrap=True)

# Group the countries with fewer than 100 movies or series under the "other" category and add it to the chart
plot.bar(["Other"], [other_countries.sum()], color="orange")

# Add axis labels
plot.xlabel("Countries")
plot.ylabel("Number of Movies/Series")

# Add a title to the chart
plot.title("Number of Movies and Series Broadcasted on Netflix by Country")

# Display the chart
plot.show()
```

```python
df = pd.read_csv('imdb_top_1000.csv')
dfgross = pd.read_csv('Highest Holywood Grossing Movies.csv')
```

```python
#replace , in gross column
df['Gross'] = df['Gross'].fillna(0)
df['Gross'] = df['Gross'].str.replace(r'\,', '')
df['Gross'] = df['Gross'].astype(str)
df['Gross'] = df['Gross'].astype(float)
df.head()
```

```python
#Combine star1,star2,star3 and star4 columns in one dataframe
column1 = df[['Star1','Gross']]
column2 = df[['Star2','Gross']]
column2.rename(columns = {'Star2':'Star1'}, inplace = True)
column3 = df[['Star3','Gross']]
column3.rename(columns = {'Star3':'Star1'}, inplace = True)
column4 = df[['Star4','Gross']]
column4.rename(columns = {'Star4':'Star1'}, inplace = True)
df_actors = pd.DataFrame()
df_actors = column1
df_actors = df_actors.append(column2,ignore_index=True)
df_actors = df_actors.append(column3,ignore_index=True)
df_actors = df_actors.append(column4,ignore_index=True)
```

```python
#Director-IMDB avarege graphic
df2 = df[["Director", "IMDB_Rating"]]
df2 = df.groupby("Director")["IMDB_Rating"].mean()
df2 = df2.sort_values( ascending=False)
df2_head = df2.head(14)
df2_head
```

```python
df2_head.plot(kind = 'bar')
```

```python
#Distributor - Grossing bar graphic
df2 = dfgross[["Distributor", "World Sales (in $)"]]
df2 = dfgross.groupby("Distributor")["World Sales (in $)"].mean()
df2 = df2.sort_values(ascending=False)
df2_head = df2.head(14)
df2_head
df2_head.plot(kind = 'bar')
plt.ylabel("World sales in hundered million")
plt.xlabel("Distributor")
```

```python
#Genres-Grossing bar graphics
from pandas import Series, DataFrame

df3 = dfgross[["Genre", "World Sales (in $)"]]


s = df3['Genre'].str.split(',').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'Genre'
del df3['Genre']
dt = df3.join(s)
dt.head()
dt = dt.groupby("Genre")["World Sales (in $)"].mean()
dt.plot(kind="bar")
#piechart
(dt.groupby("Genre")["Gross"].mean()/1000000).plot(kind = 'pie', autopct = "%1.f%%",figsize=(12,12))
#bar graph
dt = dt.groupby("Genre")["Gross"].mean()/1000000
dt =dt.sort_values(ascending=False)
dt.plot(kind="bar")
plt.title("Average grosses of IMDB top 1000 movie genres")
plt.ylabel("Total Gross (million $)")
```

```python
#publisher-grossing bar graphic
df2 = dfgross[["Distributor", "World Sales (in $)"]]
df2 = dfgross.groupby("Distributor")["World Sales (in $)"].sum()/1000000000
df2 = df2.sort_values(ascending=False)

df2_head = df2.head(14)

df2_head.plot(kind = 'bar',)
plt.ylabel("World sales(billion $)")
plt.xlabel("Distributor")
plt.title("Publisher's total revenues")

```

```python
#Count actors and make bar graphic
actors = df_actors.groupby("Star1").size()
actors = actors.sort_values(ascending=False)
actors_head = actors.head(10)
plt.bar(actors_head.index, actors_head.values)
plt.xticks(rotation=90, ha="right", wrap=True)
plt.title("Most starred actors in IMDB's top 1000 movies")
plt.ylabel("Number of films")
```

```python
#gross-actor bar graphic
df2 = df_actors.groupby("Star1")["Gross"].sum()/1000000
df2 = df2.sort_values(ascending=False)

df2_head = df2.head(20)
df2_head.plot(kind="bar")
plt.title("Total gross of IMDB top 1000 movie actors")
plt.ylabel("Total Grossing (million $)")
plt.xlabel("Actor")
```
